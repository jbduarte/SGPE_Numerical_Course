```python
import numpy as np
import matplotlib.pyplot as plt

# lets define the revenue function
def r(t):
     return -t**2+t

# create tax domain
t_domain = np.linspace(0, 1, 10**4) #the more points one creates, the better the precision one gets

# lets take a look at the function. Plot it!
plt.plot(t_domain, r(t_domain))
plt.show()
```

```python
# in this case it easy to visualize where the maximum takes place.
# lets do brute force grid search for the maximum

y = r(t_domain) # computes the Revenue for each tax rate

maximum, index = y.max(), y.argmax() # finds the maximum revenue and the index

argmax  = t_domain[index]

maximum, index, argmax
```
